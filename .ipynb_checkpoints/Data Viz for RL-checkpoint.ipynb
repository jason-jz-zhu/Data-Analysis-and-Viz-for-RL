# 1. Load and Clean Data

```python
# import Reinforcement Learning data
rl = pd.read_csv('log-250-training.txt', sep=" ", header=None)
rl = rl.iloc[:, [7, 9, 12, 14]]
rl.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import greedy data
greedy = pd.read_csv('log-greedy-best-first-search.txt', sep=" ", header=None)
greedy = greedy.iloc[:, [7, 9, 12, 14]]
greedy.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import random data
random = pd.read_csv('log-random-walk.txt', sep=" ", header=None)
random = random.iloc[:, [7, 9, 12, 14]]
random.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import a star
a_star = pd.read_csv('log-a-star.txt', sep=" ", header=None)
a_star = a_star.iloc[:, [7, 9, 12, 14]]
a_star.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# # import random blocked
random_blocked = pd.read_csv('log-random-walk-blocked-walls.txt', sep=" ", header=None)
random_blocked = random_blocked.iloc[:, [7, 9, 12, 14]]
random_blocked.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import rule based
rule_based = pd.read_csv('log-rule-based-search.txt', sep=" ", header=None)
rule_based = rule_based.iloc[:, [7, 9, 12, 14]]
rule_based.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import rule based
value_iteration = pd.read_csv('log-value-iteration.txt', sep=" ", header=None)
value_iteration = value_iteration.iloc[:, [7, 9, 12, 14]]
value_iteration.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
# import Heuristic R.L.
Heuristic_R_L = pd.read_csv('log-heuristics-reinforcement-learning.txt', sep=" ", header=None)
Heuristic_R_L = Heuristic_R_L.iloc[:, [7, 9, 12, 14]]
Heuristic_R_L.columns = ['episode_number', 'Reward', 'time_step', 'global_time']
```

```python
# clean data 
rl = rl.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
greedy = greedy.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
random = random.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
# a_star = a_star.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
random_blocked = random_blocked.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
rule_based = rule_based.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
value_iteration = value_iteration.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
Heuristic_R_L = Heuristic_R_L.applymap(lambda x: x if len(x.split(':')) == 1 else int(x.split(':')[1]))
```

```python
min_episode = min(len(rl), len(greedy), len(random), len(rule_based), len(value_iteration), len(Heuristic_R_L))
print(min_episode)
min_episode = 2000
```

```python
random_blocked.head()
```

```python
rl['Category'] = 'Reinforcement'
greedy['Category'] = 'Greedy'
# random['Category'] = 'Random Walk'
# a_star['Category'] = 'A* Search'
random_blocked['Category'] = 'Random'
rule_based['Category'] = 'Rule-based'
value_iteration['Category'] = 'Value Iteration'
Heuristic_R_L['Category'] = 'Heuristic R.L.'
```

```python
data_full = Heuristic_R_L.append([rl, greedy, random, random_blocked, rule_based, value_iteration], ignore_index=True)
data = Heuristic_R_L[:min_episode].append([rl[:min_episode], greedy[:min_episode], random[:min_episode], random_blocked[:min_episode], rule_based[:min_episode], value_iteration[:min_episode]], ignore_index=True)

print(len(data))
```

## 2.4 Heatmap

### Heatmap

### Based on the heatmap, we can find that rl has biggest count for highest level socre (>=190)

```python
def score_category(Reward):
    if Reward < -150:
        return '< -150'
    elif -150 <= Reward < -100:
        return '-150 ~ -100'
    elif -100 <= Reward < -50:
        return '-100 ~ -50'
    elif -50 <= Reward < 0:
        return '-50 ~ 0'
    elif 0 <= Reward < 50:
        return '0 ~ 50'
    elif 50 <= Reward < 100:
        return '50 ~ 100'
    elif 100 <= Reward < 150:
        return '100 ~ 150'
    elif 150 <= Reward < 190:
        return '150 ~ 190'
    elif Reward >= 190:
        return '>= 190'
```

```python
data['Reward_Category'] = data['Reward'].apply(score_category)
```

```python
tmp = data[['episode_number', 'Category', 'Reward_Category']].groupby(by=['Category', 'Reward_Category']).episode_number.count()
```

```python
columns = ['< -150', '-150 ~ -100', '-100 ~ -50', '-50 ~ 0', '0 ~ 50', '50 ~ 100', '100 ~ 150', '150 ~ 190', '>= 190']
```

```python
pivot_table = tmp.unstack(level=-1)
```

```python
heat_map_data = pivot_table.fillna(0).astype(np.int64)
```

```python
heat_map_data = heat_map_data.loc[:, columns].fillna(0).astype(np.int64)
heat_map_data = heat_map_data.reindex(['Heuristic R.L.', 'Reinforcement', 'Greedy', 'Random', 'Rule-based', 'Value Iteration'])
heat_map_data
```

```python
sns.set()
sns.set(font_scale=1.6)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(heat_map_data, annot=True, fmt='d', linewidths=.5, ax=ax, cmap=cmap, square=True, cbar_kws={'shrink': .5})
plt.xticks(rotation=25)
plt.yticks(rotation=0)
plt.show()
```
